```python
def getVectors(dataframe):
    """
    getVectors calculates the percentiles 1 through 100 for data in each columns of dataframe.
    For categorical data, the mode is used.
    :param dataframe: pandas dataframe object
    :return: pandas dataframe object with percentiles
    """
    # ensure dataframe is pandas dataframe object
    assert isinstance(dataframe, pd.DataFrame), 'Supports pandas dataframe, current class type is {}'.format(type(dataframe))
    # calculate the percentiles for numerical data and use the mode for categorical, string data
    allresults = dataframe.describe(percentiles=np.linspace(0.01, 1, num=100), include=[np.number, 'category', 'O'])
    # Pull out the percentiles for numerical data otherwise use the mode
    tempVec = allresults.apply(lambda x: x['top'] if math.isnan(x['mean']) else x.filter(regex='[0-9]{1,2}\%',
                                                                                         axis=0), axis=0)
    # test to make sure there are 100 rows
    assert tempVec.shape[0] == 100, 'Should be 100 percentiles for the data, currently returning'\
                                                    '{} rows'.format(tempVec.shape[0])

    # ensure the same amount of columns are being returned as was originally inserted into the data
    assert tempVec.shape[1] == dataframe.shape[1], 'Check function - unequal number of columns being returned'\
                                                    '\nOriginal number of columns: {}'\
                                                    '\nReturning number of columns: {}'.format(dataframe.shape[1],
                                                                                               tempVec.shape[1])
    return(tempVec)

```

```python
#Parameters
#data - [Pandas Data Frame] containing all the columns
#dependentVar - [String] Name of dependent variable
#cont_independentVar - [list] All the continuous independent variables used to build the model
#cat_independentVar - [list] All the categorical independent variables used to build the model
#modelObject - [Object] Model Object
#outputPath - [String] Path pointing to the location where .html file will be saved
#chartTitle - [String] Title of the chart
#Default
#groupByVar - [List] Plot will be shown for each level of these variables
#featureDict - [Dictionary] Mapping dictionary for variable names
#cont_incrementalVal - [List] List of length equal to the number of continuous indepedent variables containing values by which each of the variable is to be incremented


def sensitivity_plot(data,dependentVar,
                               modelObject,ouputPath,chartTitle,groupByVar = [],featureDict= {},
                            cont_incrementalVal=[],cat_incrementalVal=[]):
    
    warningMsg = ""
    try:
        if not (dependentVar in data.columns.values.tolist()):
            warningMsg = warningMsg + dependentVar+" not present in the data set."
        cont_independentVar, cat_independentVar = dummy_join(data, depVar = dependentVar)
        combineVar = cont_independentVar + cat_independentVar
        
        
        
        xTestData=data[combineVar]
        yTestData = data[dependentVar]
        yTestData = pd.to_numeric(yTestData, downcast='float')
        cont_inc_flag = True
        #Check if incrementalVal has sufficient value
        
        if len(featureDict) > 0 :
            disp_cont = list(set(featureDict) & set(cont_independentVar))
        else:
            disp_cont = cont_independentVar
        
        #if feature_dict is specified, only show those variables 
        if len(featureDict) > 0 :
            disp_cat = list(set(featureDict) & set(cat_independentVar))
        else:
            disp_cat = cat_independentVar
            
        if len(cont_incrementalVal) == 0:
            cont_inc_flag = False
            cont_incrementalVal = xTestData[disp_cont].std()
        elif not (len(cont_incrementalVal) == len(disp_cont)):
            cont_inc_flag = False
            warningMsg = "insufficient no. of units, hence taking the standard deviation"
            cont_incrementalVal = xTestData[disp_cont].std()
            
        cat_inc_flag = True
        #Check if incrementalVal has sufficient value
        if len(cat_incrementalVal) == 0:
            cat_inc_flag = False
            cat_incrementalVal_intm = xTestData[disp_cat].mode()
            for k in range(0,len(disp_cat)):
                cat_incrementalVal.extend([cat_incrementalVal_intm[disp_cat[k]][0]])
        elif not (len(cat_incrementalVal) == len(disp_cat)):
            cat_inc_flag = False
            warningMsg = "insufficient no. of units, hence taking the standard deviation2"
            cat_incrementalVal_intm = xTestData[disp_cat].mode()
            for k in range(0,len(disp_cat)):
                cat_incrementalVal.extend([cat_incrementalVal_intm[disp_cat[k]][0]])
     
        varList = cont_independentVar
        #if categorical variable is present then dummy variables are created for each level of the categorical variable, and added to varlist
        if(len(cat_independentVar) != 0):
            xTemp = pd.get_dummies(xTestData[cat_independentVar].applymap(str),prefix=cat_independentVar)
            xTestData = xTestData[cont_independentVar].join(xTemp)
            varList = varList + list(xTemp.columns)
               
        predictedYData = pd.DataFrame()
        dataString = "["
        #If there is no variable in groupByVar,
        if len(groupByVar)==0:
                groupByVar[0]="Across_Data"
                data[groupByVar[0]] = "Responsiveness"
        else:
            #Iterate for each of the groupByVar
            for y in range(0,len(groupByVar)):
                ### Jupyter - "<type 'str'>" ; Spyder - "<class 'str'>"
                ### No. levels of group by variable is hard coded as 16 for NLS Data
                #Check if continuous variable is sent as a group by variable, if yes, then bin that variable
                if(((str(type(data[groupByVar[y]].iloc[0])) != "<class 'str'>") & (str(type(data[groupByVar[y]].iloc[0])) != "<type 'str'>")) & (len(data[groupByVar[y]].unique()) > 16)):
                    binned_data= pd.qcut(data[groupByVar[y]],q = 3,labels = ["Bin_1","Bin_2","Bin_3"])
                    data[groupByVar[y]] = binned_data
        

        
        for k in range(0,len(disp_cont)):
            print(k)
            f = disp_cont[k]
            syntheticXData  = xTestData[varList]
            syntheticXData[f].add(cont_incrementalVal[k])
            predictedYData[f] = modelObject.predict(syntheticXData)
            #calculate sensitivity
            sensitivity = yTestData - predictedYData[f].tolist()
            predictedYSmooth= []
            independentVarPlot = []
            groupByVarPlot = []
            groupByVarName = []
            for y in range(0,len(groupByVar)):
                for l in data[groupByVar[y]].unique():
                    #get the indices, where this groupByVar is equal to this level
                    ind = data[data[groupByVar[y]] == l].index.tolist()
                    #Create a subset of error using the above indices
                    predictedYDataSubset = sensitivity.ix[ind]
                    #Create a subset of synthetic data using the above indices
                    syntheticXDataSubset = syntheticXData.ix[ind]
                    #Smoothing the predicted y 
                    #smooth_data=sm.nonparametric.lowess(endog = predictedYDataSubset,
                                                    #exog = syntheticXDataSubset[f],return_sorted=False,frac=alpha)
                    #smooth_data = np.where(np.isnan(smooth_data), "null", smooth_data)
                    #append the data in each iteration to create final data frame
                    predictedYSmooth.extend(list(predictedYDataSubset))
                    independentVarPlot.extend(list(syntheticXDataSubset[f]))
                    groupByVarPlot.extend(list(data.ix[ind][groupByVar[y]]))
                    groupByVarName.extend(list([groupByVar[y]]*len(predictedYDataSubset)))
            
            #create a dataframe using the list appended at each iteration
            df_temp = pd.DataFrame()
            
            df_temp['predictedYSmooth'] = pd.Series(predictedYSmooth)
            
            df_temp[f] = pd.Series(independentVarPlot)
            df_temp['groupByVarName'] = pd.Series(groupByVarName)
            df_temp['groupByValue'] = pd.Series(groupByVarPlot)
            df_temp = df_temp.fillna("null")
            df = pd.DataFrame()
            #For each level of groupby variables,remove duplicates
            for t in df_temp['groupByValue'].unique():
                df_sub_ind = df_temp[df_temp['groupByValue'] == t].index.tolist()
                df_sub = df_temp.iloc[df_sub_ind]
                df_sub_1 = df_sub.groupby([f,'predictedYSmooth']).first().reset_index()
                
                ### Split dataframe into 100 vectors for html plotting
                if len(df_sub_1) >200:
                    df_sub_1 = getVectors(df_sub_1)
                
                df = pd.concat([df_sub_1,df])
                
            #creating the feature printable list
            if f in featureDict:
                cols = list([featureDict[f]])
            else:
                cols = list([f])
            cols.extend([str(x) for x in list(df.columns[1:])])
            df = df.as_matrix()
            
            #Creating the data List for the html plotting 
            if(cont_inc_flag):
                subList='{"Type":"Continuous","Change":'+str(cont_incrementalVal[k])+',"Data":['
            else:
                #if incremental value is not passed, then by default 1 std. deviation is used as incremental value
                subList='{"Type":"Continuous","Change":"Default:'+str(cont_incrementalVal[k])+'","Data":['
            count =0
            if len(df) > 0:
                for i in df:
                    subList= subList+'{'
                    for j in range(0,len(i)):
                        #column 2,3 may contain string values, hence adding quotes
                        if j in [2,3]:
                            subList = subList+'"'+str(cols[j])+'"'+':"'+str(df[count][j])+'",'
                        else:
                            subList = subList+'"'+str(cols[j])+'"'+':'+str(df[count][j])+','
                    subList = subList[0:(len(subList)-1)] +'},'
                    count+=1
                subList = subList[:len(subList)-1]
            subList+=']},'     
            dataString+= subList
            
        
        #For insights - uncomment below
        mse = []
        groupvar = []
        grouplevel = []
        counts=[]
        for y in range(0,len(groupByVar)):
            for l in data[groupByVar[y]].unique():
                ind = data[data[groupByVar[y]] == l].index.tolist()
                grouplevel.extend([l])
                groupvar.extend([groupByVar[y]])               
                counts.append(len(data[data[groupByVar[y]] == l]))
                mse.append(mean_squared_error(yTestData.iloc[ind], predictedYData[f].iloc[ind]))
        
        
        
        df = pd.DataFrame()
        df["groupByVarName"] = pd.Series(groupvar)    
        df["groupByValue"] = pd.Series(grouplevel)   
        df["MSE"] = pd.Series(mse)
        df["Total"] = pd.Series(counts)
        cols = [str(x) for x in list(df.columns)]
        df = df.as_matrix()
        subList='{"Type":"Accuracy","Data":['
        count =0
        if len(df) > 0:
            for i in df:
                subList= subList+'{'
                for j in range(0,len(i)):
                    if j in [0,1]:
                        subList = subList+'"'+str(cols[j])+'"'+':"'+str(df[count][j])+'",'
                    else:
                        subList = subList+'"'+str(cols[j])+'"'+':'+str(df[count][j])+','
                subList = subList[0:(len(subList)-1)] +'},'
                count+=1
            subList = subList[:len(subList)-1]
        subList+=']},'     
        dataString+= subList
        

        
        for k in range(0,len(disp_cat)):
            print(len(disp_cont)+k)
            f = disp_cat[k]
            
            syntheticXData  = xTestData[varList]
            #incremental change using the cat_incremental value
            for elem in data[disp_cat[k]].unique():
                    if(elem == cat_incrementalVal[k]):
                        syntheticXData[disp_cat[k]+'_'+str(elem)] = 1
                    else:
                        syntheticXData[disp_cat[k]+'_'+str(elem)] = 0
            predictedYData[f] = modelObject.predict(syntheticXData)
            #calculate sensitivity
            sensitivity = predictedYData[f] - yTestData.tolist()
            predictedYSmooth= []
            independentVarPlot = []
            groupByVarPlot = []
            groupByVarName = []
            syntheticXData[f] = data[disp_cat[k]]
            
            for y in range(0,len(groupByVar)):
                for l in data[groupByVar[y]].unique():
                    #get the indices, where this groupByVar is equal to this level
                    ind = data[data[groupByVar[y]] == l].index.tolist()
                    predictedYDataSubset = sensitivity.ix[ind]
                    syntheticXDataSubset = syntheticXData.ix[ind]
                    #Iterate for eac unique value of ategorical independent variable
                    for m in syntheticXDataSubset[disp_cat[k]].unique():
                        #if level m is equal to cat_incremental value for this variable, then skip
                        if m == cat_incrementalVal[k]:
                            continue
                        #get the indices where this categorical independent variable is equal to this unique value
                        ind_cat = syntheticXDataSubset[syntheticXDataSubset[f] == m].index.tolist()
                        predictedYSmooth.extend([predictedYDataSubset.ix[ind_cat].mean()])
                        #append unique value of categorical independent variable
                        independentVarPlot.extend([m])
                        #append level of groupby variable
                        groupByVarPlot.extend([l])
                        #append groupby variable
                        groupByVarName.extend([groupByVar[y]])
                
            #create a dataframe using the above lists created by appending for each iteration of unique values
            df_temp = pd.DataFrame()
            
            df_temp['predictedYSmooth'] = pd.Series(predictedYSmooth)
            
            df_temp[f] = pd.Series(independentVarPlot)
            df_temp['groupByVarName'] = pd.Series(groupByVarName)
            df_temp['groupByValue'] = pd.Series(groupByVarPlot)
            #replace na with null values
            df_temp = df_temp.fillna("null")
            df = pd.DataFrame()
            #For each level of groupby variables,keep only first observation
            for t in df_temp['groupByValue'].unique():
                df_sub_ind = df_temp[df_temp['groupByValue'] == t].index.tolist()
                df_sub = df_temp.iloc[df_sub_ind]
                df_sub_1 = df_sub.groupby([f,'predictedYSmooth']).first().reset_index()
                df = pd.concat([df_sub_1,df])
            #creating the feature printable list
            if f in featureDict:
                cols = list([featureDict[f]])
            else:
                cols = list([f])
            cols.extend([str(x) for x in list(df.columns[1:])])
            df = df.as_matrix()
            
            #Creating the data List for the html plotting 
            if(cat_inc_flag):
                subList='{"Type":"Categorical","Change":"'+cat_incrementalVal[k]+'","Data":['
            else:
                subList='{"Type":"Categorical","Change":"Default:'+cat_incrementalVal[k]+'","Data":['
            count =0
            if len(df) > 0:
                for i in df:
                    subList= subList+'{'
                    for j in range(0,len(i)):
                        #columns 0,2,3 may have string values, hence adding quotes in order to process it in d3
                        if j in [0,2,3]:
                            subList = subList+'"'+str(cols[j])+'"'+':"'+str(df[count][j])+'",'
                        else:
                            subList = subList+'"'+str(cols[j])+'"'+':'+str(df[count][j])+','
                    subList = subList[0:(len(subList)-1)] +'},'
                    count+=1
                subList = subList[:len(subList)-1]
            subList+=']},'    
            dataString+= subList    
        dataString = dataString[:len(dataString)-1]
        dataString+="]"  
        #Composing and creating the html plot file 
        htmlcode= getMLImpactHtmlCode(dataString,dependentVar)
        
        htmlfileName=ouputPath+"/MLImpact.html"   
        # Generating a scatter plot as a html file 
        
        f = open(htmlfileName,'w')
        
        f.write(htmlcode)
        f.close()
        warningMsg = warningMsg + "HTML file generated at the given path"
    except Exception as e:
        warningMsg = e
        pass
    return warningMsg
```

```python
#method to split data into cateogrical and continuous variable lists for later use 
#also creates necessary dummies for modeling 
def dummy_join(final_data, depVar):
    
    #split the dataset based on whether it is of the object type or not into continuous and categorical variables
    categorical = final_data.select_dtypes(include={'object'})
    continuous =  final_data.select_dtypes(exclude={'object'})

    cat_independentVar = list(categorical.columns)
    #make the categorical data into dummies and get those columns
    xTemp = pd.get_dummies(categorical.applymap(str), prefix = categorical.columns, dummy_na = True)

    cat_dummy = list(xTemp.columns)

    #get the column list of the continuous data
    cont_independentVar = list(continuous.columns)
    try:
        cont_independentVar.remove(depVar)
    except:
        pass
    
    return  cont_independentVar, cat_independentVar
    #join the base continuous variables to the dummies from the categorical data to get a final dataset
    #that is suitable for modeling 
    #xTrainData = continuous.join(xTemp).join(categorical)
#cont_independentVar2, cat_independentVar2 = dummy_join(data, depVar = dependentVar)
```
